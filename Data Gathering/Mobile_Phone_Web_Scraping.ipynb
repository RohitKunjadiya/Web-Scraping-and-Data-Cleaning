```python
# Load html file
with open('smartprix.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
# Fetch the webpage and parse it using BeautifulSoup
from bs4 import BeautifulSoup
# Import necessary libraries
import numpy as np
# Import necessary libraries
import pandas as pd
```

```python
# Fetch the webpage and parse it using BeautifulSoup
soup = BeautifulSoup(html,'lxml')
```

```python
# Extract specific elements from the HTML
containers = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
names = []
prices = []
ratings = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in containers:
    try:
    # Extract specific elements from the HTML
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
    # Extract specific elements from the HTML
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
    try:
    # Extract specific elements from the HTML
        ratings.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        ratings.append(np.nan)
        
    # Extract specific elements from the HTML
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:    
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
            
```

```python
# Convert the scraped data into a Pandas DataFrame
df = pd.DataFrame({
    'model':names,
    'price':prices,
    'rating':ratings
})
```
